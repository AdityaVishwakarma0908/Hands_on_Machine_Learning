```python
import pandas as pd
import joblib
import numpy as np

# Load the saved files
model = joblib.load("california_housing_model.pkl")
pipeline = joblib.load("full_pipeline.pkl")
```

```python
# Raw data for a new house (Notice we include the 'ocean_proximity' text)
new_house_data = pd.DataFrame([{
    "longitude": -150.24,
    "latitude": 35.05,
    "housing_median_age": 30,
    "total_rooms": 2500,
    "total_bedrooms": 450,
    "population": 120,
    "households": 400,
    "median_income": 1.5,
    "ocean_proximity": "<1H OCEAN" # The model needs this to create the One-Hot columns!
}])
```

```python
new_data_prepared = pipeline.transform(new_house_data)

prediction = model.predict(new_data_prepared)

print(prediction)
```
